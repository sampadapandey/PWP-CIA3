# Business Understanding 

Attrition is a reduction of staff for voluntary or involuntarily reasons. Every year a lot of companies hire a number of employees. The companies invest time and money in training those employees, not just this but there are training programs within the companies for their existing employees as well. The aim of these programs is to increase the effectiveness of their employees. It is a challenging task for any organization to retain its employees as hiring new employee and conducting the whole recruiting process requires a lot of spending in terms of time and money. Attrition refers to the turnover rate of employees inside an organization. Therefore, the issue of keeping one's employees happy and satisfied is an age-old challenge. In order to reduce the rate of attrition inside any organization, we need to study the factors responsible for employee attrition so the organization can focus on them and thereby avoid the negative effect of attrition. 
Human resource analytics is an area in the field of analytics that refers to applying analytic processes to the human resource department of an organization in the hope of improving employee performance and therefore getting a better return on investment. HR analytics does not just deal with gathering data on employee efficiency.It helps us with interpreting organizational data. It finds out the people-related trends in the data and helps the HR Department take the appropriate steps to keep the organization running smoothly and profitably. Attrition is a corporate setup is one of the complex challenges that the people managers and the HRs personnel have to deal with the. Instead, it aims to provide insight into each process by gathering data and then using it to make relevant decisions about how to improve these processes.

The aim of our research was therefore to (1) handle missing values (2) dealing with outliers (3) establishing some correlation coefficient between listed factors and whether an employee is going to quit or not.

# Finding correlation for all numeric variables

From the above heat map we can now see which variables are poorly correlated and which ones are strongly correlated. Some insights that can be drawn are-
1.JobLevel is highly related to Age as expected as aged employees will generally tend to occupy higher positions in the company.
2.MonthlyIncome is very strongly related to joblevel as expected as senior employees will definately earn more.
3.PerformanceRating is highly related to PercentSalaryHike which is quite obvious.
4.TotalWorkingYears is highly related to JobLevel which is expected as senior employees must have worked for a larger span of time.
5.YearsWithCurrManager is highly related to YearsAtCompany.
6.YearsAtCompany is related to YearsInCurrentRole.

# Plotting the features against the 'Target' variable

```python
sns.catplot(data=data,kind='count',x='Attrition',col='Department')
```

R&D department records maximum attrition of employees.

```python
sns.catplot(data=data,kind='bar',x='Attrition',y='MonthlyIncome')
```

The average income for employees not quitting the organisation is quite higher and it is obvious as those earning well will certainly not be willing to leave. Similarly those employees who are probably not earning well will certainly want to change the company.

```python
plt.figure(figsize=(20,10))
sns.countplot(data['Age'],hue=data['Attrition'])
plt.show()
```

We can observe that from age 0-29 max people are leaving the company. This shows that people with higher age have lesser tendency to leave the organisation which makes sense as they may have settled in the organisation.

```python
plt.figure(figsize=[10,20])

plt.subplot(411)
sns.countplot(x=data['JobSatisfaction'],hue=data['Attrition'])
plt.subplot(412)
sns.countplot(x=data['MaritalStatus'],hue=data['Attrition'])
plt.subplot(413)
sns.countplot(x=data['JobRole'],hue=data['Attrition'])
plt.subplot(414)
sns.countplot(x=data['JobLevel'],hue=data["Attrition"])
plt.show()
```

This concludes that Single employees tend to leave as compared to married and divorced. Sales Representitives tend to leave compared to any other job and Less involved employees tend to leave the company. Less experienced (low job level) tend to leave the company.
This maybe because of reasons like married employees prefer stable jobs whereas single employees don't mind relocating their jobs. Sales representatives might exprience lesser pay and bonus according to the task they are performing hence, they tend to leave their job and less involved and low job level employees tend to leave organisation for a better job offer and pay scale.

```python
sns.barplot(data=data, x = 'Attrition', y = 'DistanceFromHome')
```

This denotes that People are more likely to quit When 'DistanceFromHome' is between 9 to 12

```python
sns.countplot(data['Gender'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Gender')
```

The bar graph denotes that count of male employees is higher than that of females. 

```python
fig=plt.figure(figsize=(8,10))
sns.countplot(x='TotalWorkingYears',hue='Attrition',data=data ,color='orange')
plt.show()
```

Employees with TotalWorkingYears of 11 to 13 are less likely to quit their job. 

```python
sns.countplot(data['BusinessTravel'])
fig = plt.gcf()
fig.set_size_inches(8,8)
plt.title('Business travel')
```

This inference tells that employees who got rare opportunity to travel are likely to quit their job as they weren't satisfied with their job and the learnings or exposure they were getting.
